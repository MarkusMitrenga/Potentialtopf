# Anwendung des Potentialtopf-Modells

## Die gefundenen Gleichungen
Für den linearen Potentialtopf können mit Hilfe der Schödinger-Gleichung die folgenden Aussagen hergeleitet werden:
* Gesamtenergie $E_{ges}$

$E_{ges} = \frac{\hbar^{2} \pi^{2}}{2m_{e}L^{2}}\cdot n^{2}$ mit $n = 1,2,3,...$ 

Hierbei ist n die **Quantenzahl** der Gesamtenergie, die sich daher quantisiert ändert.

* Wellenfunktion $\psi(x)$

$\psi(x) = \sqrt{\frac{2}{L}} \cdot sin\left(\frac{n \pi}{L} \cdot x \right)$

Nach **Born** lässt sich aus der Wellenfunktion die Wahrscheinlichkeit gewinnen, mit der das Elektron an einem bestimmten Platz im Potentialtopf zu finden ist: $P(x) = |\psi(x)|^{2}$

Da die Wahrscheinlichkeit das Elektron *irgendwo* im Potentialtopf zu finden 1 beträgt mus gelten:
$\int\limits_{0}^{L} |\psi(x)|^{2} \ dx \stackrel{!}{=} 1$

1. *Bestimme* mit Hilfe des Skriptes unten die Breite des Potentialtopfes $L$, Gesamtenergie $E_{ges}$, die Wellenfunktion $\psi(x)$ und die Wahrscheinlichkeitsdichte $|\psi(x)|^{2}$ für $n=1$. Zur Ermittlung der Breite des Potentialtopfes muss man die rot markierten Bindungen zählen. Diese Anzahl multipliziert mit einer mittleren Bindungslänge, die in dem Skript bereits enthalten ist, ergibt dann $L$.
2. *Bestimme* aus dem Diagramm zur Wahrscheinlichkeitsdichte den Wert von $\int\limits_{0}^{L} |\psi(x)|^{2} \ dx$. *Erläutere*, ob dieser Wert plausibel ist.
3. Wiederhole die Arbeitsschritte unter (1) für $n=2$, $n=3$ und $n=4$. *Beobachte* die Änderung der Gesamtenergie und der Wellenfunktion $\psi(x)$. Wie kann man an der Wellenfunktion erkennen, dass sie zu einem höheren Wert der Gesamtenergie gehört?

```python
#Imports
import numpy as np
import pylab as plt

#Groesse der Abbildung
plt.rcParams['figure.figsize'] = (12, 8)

#Konstanten
h = 6.62607004e-34
me = 9.10938356e-31
l = 140e-12
c = 299792458
korr=3e-19

#Anzahl der Bindungen
b = float(input('Anzahl der Bindungen angeben: '))
anzDB = (b-1)/2

#Breite Potentialtopf
a = b * l

#Quantenzahl n
n = int(input('Energiequantenzahl n: '))

#Berechnung Energien und psi
L = int(a * 10e12) #Breite Potentialtopf
X = np.linspace(0, L, L, endpoint=True) #Wertebereich
psi1 = np.sqrt(2/L)*np.sin((n*np.pi)/L *X) #Wellenfunktion
psi2 = (np.sqrt(2/L)*np.sin((n*np.pi)/L *X))**2 #Wahrscheinlichkeitsdichte
Eges = h**2/(8*me*a**2)*n**2 #Gesamtenergie

#Ausgabe Energien
print('Anzahl der konjugierten Doppelbindungen: ', int(anzDB))
print('n =', int(n),'Eges =' , '%10.3e'% (Eges) , 'J')

#Plot Wellenfunktionen
plt.subplot(121)
plt.plot(X,psi1)
plt.grid(True)
plt.title('Wellenfunktion $\psi(x)$')
plt.xlabel('L [pm]')
plt.ylabel('$\psi(x)$')
plt.subplot(122)
plt.plot(X,psi2)
plt.grid(True)
plt.title('Wahrscheinlichkeitsdichte')
plt.xlabel('L [pm]')
plt.ylabel('$|\psi(x)|^{2}$')
plt.tight_layout(pad=1, w_pad=1, h_pad=2.0)
plt.show()
```

## Deutung der Absorption mit dem Potentialtopf-Modell

Bei der Absorption von Licht werden Elektronen von einem Grundzustand in einen angeregten Zustand angehoben. Dabei werden ausschließlich die Elektronen der Doppelbindungen ($\pi$-Elektronen) berücksichtigt und jeder Energiezustand kann von zwei $\pi$-Elektronen eingenommen werden. Da 1,3-Butadien 4 $\pi$-Elektronen besitzt, sind die Energiezustände für $n = 1$ und $n = 2$ besetzt. Der nächste angeregte Zustand, in den Elektronen "springen" können, wird durch die Quantenzahl $n= 3$ charakterisiert. Das Vorgehen ist also wie folgt:

* Berechne $E_{ges}$ für n=2
* Berechne $E_{ges}$ für n=3
* Bilde die Differenz $\Delta E_{ges}$
* Rechne mit Hilfe der Gleichungen $\lambda \cdot f = c$ und $E = h \cdot f$ die Energie in die Wellenlänge $\lambda$ um.

Da die $\pi$-Elektronen in den Doppelbindungen nicht vollständig beweglich sind, muss zu der gefundenen Energiedifferenz $\Delta E_{ges}$ noch ein Korrekturwert $E_{korr} = 3 \cdot 10^{-19} J$ addiert werden.

### $\beta$-Carotin

2. *Bestimme* das zu erwartende Absorptionsmaximum für $\beta$-Carotin (s.o.) und vergleiche mit dem real gemessenen Spektrum (Skript weiter unten). Welche Wellenlänge/welcher Farbbereich wird durch $\beta$-Carotin absorbiert? Begründe.

```python
import numpy as np

#Konstanten
h = 6.62607004e-34
me = 9.10938356e-31
l = 140e-12
c = 299792458
korr=3e-19

#Anzahl der Bindungen
b = float(input('Anzahl der Bindungen angeben: '))
anzDB = (b-1)/2
print('Die Anzahl der Doppelbindungen beträgt:', int(anzDB))

#Breite Potentialtopf
a = b * l

#Quantenzahl n
n = anzDB
m = n+1
print('\nQuantenzahl für die höchste besetzte Energiestufe:',int(n))
print('Quantenzahl für die niedrigste unbesetzte Energiestufe:',int(m))

#Berechnung Energien und psi
L = int(a * 10e12) #Breite Potentialtopf
Eges_besetzt = h**2/(8*me*a**2)*n**2 #Gesamtenergie
Eges_unbesetzt = h**2/(8*me*a**2)*m**2
deltaE = Eges_unbesetzt-Eges_besetzt
korr_deltaE = deltaE + korr

#Berechnung Lambda
lamb = (c*h)/korr_deltaE *1e9

#Ausgabe Energien
print('\nn =', int(n),'Eges =' , '%10.3e'% (Eges_besetzt) , 'J')
print('n =', int(m),'Eges =' , '%10.3e'% (Eges_unbesetzt) , 'J')
print('Delta E =' , '%10.3e'% (deltaE) , 'J')
print('Korrigiertes Delta E =' , '%10.3e'% (korr_deltaE) , 'J')

#Ausgabe Wellenlaenge
print('\nWellenlänge lambda =' , '%4.0f'% (lamb) , 'nm')
```
